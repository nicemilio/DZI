## Lösung 1.1: Pearson-Korrelation

**Gegebene Daten:**

## Lösung 1.2: Lineare Regression mit Methode der kleinsten Quadrate

**Formeln für die Methode der kleinsten Quadrate:**

$$\omega_1 = \frac{\sum_{i=1}^{n}(t_i - \bar{t})(s_i - \bar{s})}{\sum_{i=1}^{n}(t_i - \bar{t})^2}$$

$$\omega_0 = \bar{s} - \omega_1 \cdot \bar{t}$$

$$\omega_1 = \frac{-90.5}{354} = -0.2557$$

**Berechnung von $\omega_0$:**

$$\omega_0 = 4 - (-0.2557) \cdot 17.5 = 4 + 4.4748 = 8.4748$$

**Ergebnis:**

$$\boxed{s(t) = 8.4748 - 0.2557 \cdot t}$$

## Lösung 1: Knotenarten in Entscheidungsbäumen

**1. Entscheidungsknoten (innere Knoten):**
- Testen ein Attribut/Merkmal der Eingabedaten
- Verzweigen zu verschiedenen Kindknoten basierend auf dem Testergebnis
- Jede ausgehende Kante entspricht einem möglichen Attributwert

**2. Blattknoten (terminale Knoten):**
- Enthalten die finale Vorhersage/Klassifikation
- Haben keine Kindknoten
- Bei Klassifikation: enthalten die Klassenbezeichnung
- Bei Regression: enthalten den vorhergesagten Wert (z.B. Mittelwert der Trainingsdaten im Blatt)
