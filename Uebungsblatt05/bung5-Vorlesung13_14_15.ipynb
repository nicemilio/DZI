3. **Lernen aus Daten**: Durch iteratives Anpassen der Parameter θ anhand der Trainingsdaten "lernt" das Modell die zugrundeliegenden Muster und Zusammenhänge, die zur Klassifikation benötigt werden.

**Die VC-Dimension einer Linie (in 2D) ist 3.**

**Begründung:**
- Die VC-Dimension gibt an, wie viele Punkte eine Hypothesenklasse maximal "zerschmettern" (shatter) kann, d.h. alle möglichen 2^n Labelkombinationen realisieren kann.

- **3 Punkte können zerschmettert werden:** Für 3 Punkte in allgemeiner Lage (nicht kollinear) kann eine Linie jede der 2³ = 8 möglichen Labelkombinationen (+/- für jeden Punkt) durch entsprechende Positionierung trennen.

- **4 Punkte können NICHT immer zerschmettert werden:** Bei 4 Punkten gibt es immer eine Konfiguration (z.B. XOR-Anordnung, bei der diagonal gegenüberliegende Punkte das gleiche Label haben), die nicht durch eine einzelne Linie getrennt werden kann.

Daher: **VC-Dimension = 3**

4. Formulieren Sie den Satz von Bayes und erklären Sie dabei die auftretenden Komponenten in eigenen Worten. (4 Punkte)

**Satz von Bayes:**

$$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$$

**Komponenten:**

- **P(A|B) – Posterior (A-posteriori-Wahrscheinlichkeit):**
  Die Wahrscheinlichkeit der Hypothese A, nachdem die Evidenz B beobachtet wurde. Dies ist das, was wir berechnen wollen.

- **P(B|A) – Likelihood (Wahrscheinlichkeit der Evidenz):**
  Die Wahrscheinlichkeit, die Evidenz B zu beobachten, gegeben dass die Hypothese A wahr ist.

- **P(A) – Prior (A-priori-Wahrscheinlichkeit):**
  Die ursprüngliche Wahrscheinlichkeit der Hypothese A, bevor irgendwelche Evidenz beobachtet wurde. Repräsentiert unser Vorwissen.

- **P(B) – Evidence (Marginale Wahrscheinlichkeit):**
  Die Gesamtwahrscheinlichkeit, die Evidenz B zu beobachten, unabhängig von der Hypothese. Dient als Normalisierungskonstante.
  Kann berechnet werden als: P(B) = P(B|A)·P(A) + P(B|¬A)·P(¬A)

**Interpretation:** Der Satz von Bayes ermöglicht es, unsere Überzeugung über eine Hypothese zu aktualisieren, wenn neue Evidenz verfügbar wird.

5. Satz von Bayes (engl. Bayes' rule). Wir haben 3 Karten. Eine ist auf beiden Seiten rot, eine Weitere auf beiden Seiten schwarz. Die letzte Karte ist auf einer Seite schwarz und auf der anderen Seite rot. Die 3 Karten befinden sich vermischt in einem Hut. Eine Karte wird zufällig gezogen und auf den Boden gelegt. Wenn die Seite, die nach oben zeigt, rot ist, wie hoch ist die Wahrscheinlichkeit, dass die andere Seite schwarz ist? (3 Punkte)

**Lösung mit Satz von Bayes:**

**Karten:**
- Karte 1: Rot-Rot (RR)
- Karte 2: Schwarz-Schwarz (SS)
- Karte 3: Rot-Schwarz (RS)

**Gesucht:** P(andere Seite ist schwarz | obere Seite ist rot)

**Berechnung:**
- Es gibt insgesamt 6 Seiten, davon 3 rote Seiten
- Die 3 roten Seiten verteilen sich: 2 auf Karte RR, 1 auf Karte RS

Wenn wir eine rote Seite sehen, welche Karte haben wir gezogen?
- P(sehen rot | RR) = 1 (beide Seiten rot)
- P(sehen rot | SS) = 0 (keine rote Seite)
- P(sehen rot | RS) = 0.5 (eine von zwei Seiten ist rot)

**Mit Bayes:**

A = andere Seite ist schwarz = wir haben Karte RS gezogen
B = obere Seite ist rot

P(RS | rot oben) = P(rot oben | RS) · P(RS) / P(rot oben)

- P(RS) = 1/3
- P(rot oben | RS) = 1/2
- P(rot oben) = P(rot|RR)·P(RR) + P(rot|SS)·P(SS) + P(rot|RS)·P(RS)
- P(rot oben) = 1·(1/3) + 0·(1/3) + 0.5·(1/3) = 1/3 + 1/6 = 1/2

P(RS | rot oben) = (1/2 · 1/3) / (1/2) = (1/6) / (1/2) = **1/3**

**Antwort: Die Wahrscheinlichkeit, dass die andere Seite schwarz ist, beträgt 1/3.**

**Klassifikation von d6: N=n, C=y, R=y, F=n**

P(krank|d6) ∝ P(krank) · P(N=n|krank) · P(C=y|krank) · P(R=y|krank) · P(F=n|krank)

# Vorlesung 15: Maximum-Likelihood-Schätzung

1. Gegeben sei eine Stichprobe von Messungen: $x_1 = 5, x_2 = 12, x_3 = 17, x_4 = 5, x_5 = 8$. Nehmen Sie an, dass das Merkmal normalverteilt ist und geben Sie einen wahrscheinlichen Populationsparameter $\mu$ an. (2 Punkte)

**Gegeben:** x₁ = 5, x₂ = 12, x₃ = 17, x₄ = 5, x₅ = 8

**Annahme:** Das Merkmal ist normalverteilt.

**MLE-Schätzer für μ:**
Der Maximum-Likelihood-Schätzer für den Erwartungswert μ einer Normalverteilung ist das arithmetische Mittel der Stichprobe:

$$\hat{\mu} = \bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$$

**Berechnung:**
$$\hat{\mu} = \frac{5 + 12 + 17 + 5 + 8}{5} = \frac{47}{5} = \mathbf{9.4}$$

**Antwort: Der wahrscheinliche Populationsparameter μ ist 9.4**

2. Es sei X eine diskrete Zufallsvariable mit der folgenden Wahrscheinlichkeitsfunktion (wobei 0 ≤ θ ≤ 1 ein Parameter ist):

| X | 0 | 1 | 2 | 3 |
| --- | --- | --- | --- | --- |
| P(X) | 2$\theta$/3 | $\theta$/3 | 2(1-$\theta$)/3 | (1-$\theta$)/3 |

Die folgende Stichprobe aus 10 voneinander unabhängigen Messungen sei von dieser Verteilung gegeben: (3,0,2,1,3,2,1,0,2,1). 

Ermitteln Sie die Maximum-Likelihood-Schätzung von $\theta$. (6 points)

**Gegeben:** Stichprobe (3, 0, 2, 1, 3, 2, 1, 0, 2, 1) mit n = 10

**Zählung der Werte:**
- X = 0: 2 mal → n₀ = 2
- X = 1: 3 mal → n₁ = 3
- X = 2: 3 mal → n₂ = 3
- X = 3: 2 mal → n₃ = 2

**Likelihood-Funktion:**
$$L(\theta) = \prod_{i} P(X=x_i) = \left(\frac{2\theta}{3}\right)^{n_0} \cdot \left(\frac{\theta}{3}\right)^{n_1} \cdot \left(\frac{2(1-\theta)}{3}\right)^{n_2} \cdot \left(\frac{1-\theta}{3}\right)^{n_3}$$

$$L(\theta) = \left(\frac{2\theta}{3}\right)^{2} \cdot \left(\frac{\theta}{3}\right)^{3} \cdot \left(\frac{2(1-\theta)}{3}\right)^{3} \cdot \left(\frac{1-\theta}{3}\right)^{2}$$

**Vereinfachung (Konstanten ignorieren):**
$$L(\theta) \propto \theta^2 \cdot \theta^3 \cdot (1-\theta)^3 \cdot (1-\theta)^2 = \theta^5 \cdot (1-\theta)^5$$

**Log-Likelihood:**
$$\ell(\theta) = 5 \ln(\theta) + 5 \ln(1-\theta)$$

**Ableitung und Nullsetzen:**
$$\frac{d\ell}{d\theta} = \frac{5}{\theta} - \frac{5}{1-\theta} = 0$$

$$\frac{5}{\theta} = \frac{5}{1-\theta}$$

$$1 - \theta = \theta$$

$$1 = 2\theta$$

$$\hat{\theta} = \mathbf{0.5}$$

**Antwort: Die Maximum-Likelihood-Schätzung von θ ist 0.5**

**Parametrische Klassifikation mit MLE**

**Daten:**
- Basic: 9, 10, 5, 3, 9 (n₁ = 5)
- Advanced: 9, 5, 15, 12, 21 (n₂ = 5)

---

**Schritt 1: Prior-Wahrscheinlichkeiten**
- P(basic) = 5/10 = **0.5**
- P(advanced) = 5/10 = **0.5**

---

**Schritt 2: MLE-Schätzer für μ (Mittelwerte)**

**Basic:**
$$\hat{\mu}_{basic} = \frac{9 + 10 + 5 + 3 + 9}{5} = \frac{36}{5} = \mathbf{7.2}$$

**Advanced:**
$$\hat{\mu}_{advanced} = \frac{9 + 5 + 15 + 12 + 21}{5} = \frac{62}{5} = \mathbf{12.4}$$

---

**Schritt 3: Stichprobenvarianz σ²**

**Basic:** $(x_i - 7.2)^2$
- (9-7.2)² = 3.24
- (10-7.2)² = 7.84
- (5-7.2)² = 4.84
- (3-7.2)² = 17.64
- (9-7.2)² = 3.24
- Summe = 36.8

$$\hat{\sigma}^2_{basic} = \frac{36.8}{5-1} = \frac{36.8}{4} = \mathbf{9.2}$$
$$\hat{\sigma}_{basic} = \sqrt{9.2} = \mathbf{3.033}$$

**Advanced:** $(x_i - 12.4)^2$

- (5-12.4)² = 54.76
- (15-12.4)² = 6.76
- (12-12.4)² = 0.16

- Summe = 147.2

$$\hat{\sigma}_{advanced} = \sqrt{36.8} = \mathbf{6.066}$$

**Schritt 4: Likelihood für x = 11 Monate**

Normalverteilung: $P(x|Klasse) = \frac{1}{\sqrt{2\pi}\sigma} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$

**Ergebnis:**
P(advanced|x=11) = 0.0321 > P(basic|x=11) = 0.0301
