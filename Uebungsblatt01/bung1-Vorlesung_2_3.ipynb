### 5. Bitte erklären Sie "Logische Datenunabhängigkeit" und "Physische Datenunabhängigkeit" in der Drei-Schichten-Architektur in Ihren eigenen Worten. (3 Punkte)

Die logische Datenunabhängigkeit bedeutet, dass Änderungen am konzeptuellen Schema (z. B. neue Tabellen) keine Anpassungen der externen Sichten oder Anwendungsprogramme erfordern.
Die physische Datenunabhängigkeit bedeutet, dass Änderungen an der physischen Speicherung (z. B. Speicherort, Speichermedium) keinen Einfluss auf das konzeptuelle Schema oder die Anwendungen haben.
Beide Konzepte dienen dazu, die Datenbankarchitektur flexibel und wartungsfreundlich zu gestalten.
