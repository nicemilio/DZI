    2.1 Nennen sie die drei Arten von Zufallsstichproben und wie sie durchgeführt werden. (4 Punkte)

Einfache Zufallsstichprobe (Simple Random Sample):

Jedes Objekt der Grundgesamtheit hat die gleiche Wahrscheinlichkeit, ausgewählt zu werden. Vorge-
hensweise: Die Objekte werden in eine “Urne” (z. B. eine nummerierte Liste) platziert. Mithilfe von
Zufallszahlen werden Objekte ohne Zurücklegen ausgewählt, bis die gewünschte Stichprobengröße
erreicht ist. Anwendbarkeit: wenn über die Verteilung der relevanten Merkmale in der Grundge-
samtheit keine Informationen vorliegen.

Klumpenstichprobe (Cluster Sample):

umfasst alle Objekte, die in mehreren zufällig ausgewählten Klumpen (Clustern) enthalten sind
(z.B.: Schulklassen) Vorgehensweise: Die Grundgesamtheit wird in Cluster unterteilt. Mehrere
Cluster werden zufällig ausgewählt. Alle Objekte innerhalb der ausgewählten Cluster werden unter-
sucht. (Hinweis: ein einzelnes Cluster stellt eine Ad-hoc-Stichprobe dar) Anwendbarket: Geeignet,
wenn die Objekte in sinnvolle Gruppen (Cluster) vorstrukturiert sind.

Geschichtete Zufallsstichprobe (Stratified Random Sample):

Vorgehensweise: Die Grundgesamtheit wird in nicht überlappende Untergruppen (Schichten)
aufgeteilt, basierend auf relevanten Merkmalen. Innerhalb jeder Schicht erfolgt eine zufällige
Auswahl, z. B. mithilfe einer einfachen Zufallsstichprobe oder des Klumpenverfahrens. In einer
proportional geschichteten Stichprobe entspricht die prozentuale Verteilung der Schichtungsmerk-
male der in der Population. Anwendbarkeit: Geeignet, wenn bekannt ist, dass bestimmte Merkmale
die Verteilung beeinflussen (z. B. Alter, Einkommen) und diese in der Stichprobe repräsentiert sein sollen

    2.2 Warum eignen sich Nicht-probabilistische Stichproben in der Realität nicht für statistische Inferenz? (1 Punkt) 

    Begründen Sie mit Hilfe folgenden Beispieles:

ist verzeert und kann nicht die Merkmale der gesammten Population wieder geben.
